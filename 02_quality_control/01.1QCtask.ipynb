```r
# 根据raw QC图去做质控分析，加载包
# step0 读取包
library(Seurat, lib.loc = "/opt/conda/lib/R/library")
library(data.table, lib.loc = "/opt/conda/lib/R/library")
library(dplyr, lib.loc = "/opt/conda/lib/R/library")
library(DoubletFinder, lib.loc = "/opt/conda/lib/R/library")
library(ggplot2)
# step1 设置参数
prefix <- "cot"  # ← 在这里定义前缀，方便修改
rds <- "/data/work/RNA_AraSeed/rds/cot_merge_raw.rds"
outdir <- "/data/work/RNA_AraSeed/CellTypeAnno/cot"
## 参数按照之前的vlnplot图去确定参数
minUMIs        <- as.numeric(1000)   # 1000
maxUMIs        <- as.numeric(80000)
minGenes       <- as.numeric(500)   # 500
maxPercent.mt  <- as.numeric(1)   # 2
maxPercent.C   <- as.numeric(5)   # 5
dim.usage      <- as.numeric(20)   # 20
res.usage      <- as.numeric(0.5)  # 0.5
doublet.rate   <- as.numeric(0.05) # 0.05 这个一般都是规定的0.05
setwd(outdir)

## =========================
## 2. 读取rds
## =========================
scRNA <- readRDS(rds)
scRNA$Stage <- prefix  # 也可以使用 prefix 变量

## =========================
## 3. 过滤前查看文库状态--可根据此去调整上面的参数，可判断双胞过滤是否起作用
## =========================

pdf(paste0("QC_VlnPlot_beforeQC_", prefix, ".pdf"), width = 18, height = 5)
VlnPlot(
  scRNA,
  features = c("nCount_RNA", "nFeature_RNA", "percent.mt", "percent.C"),
  group.by = "Stage",
  pt.size = 0
)
dev.off()

## =========================
## 4. 按标准过滤细胞
## =========================
scRNA <- subset(
  scRNA,
  subset =
    nCount_RNA   >= minUMIs &
    nCount_RNA   <= maxUMIs & 
    nFeature_RNA >= minGenes &
    percent.mt   <= maxPercent.mt &
    percent.C    <= maxPercent.C
)

cat("Cells after QC:", ncol(scRNA), "\n")

## =========================
## 5. 标准降维与聚类
## =========================

scRNA <- NormalizeData(scRNA)
scRNA <- FindVariableFeatures(scRNA, nfeatures = 2000)
scRNA <- ScaleData(scRNA)
scRNA <- RunPCA(scRNA, npcs = max(dim.usage, 30))

scRNA <- FindNeighbors(scRNA, dims = 1:dim.usage)
scRNA <- FindClusters(scRNA, resolution = res.usage)
scRNA <- RunUMAP(scRNA, dims = 1:dim.usage)

## =========================
## 6. DoubletFinder 去双胞
## =========================

cat(">>> Running DoubletFinder\n")

sweep.res <- paramSweep_v3(scRNA, PCs = 1:dim.usage, sct = FALSE)
sweep.stats <- summarizeSweep(sweep.res)
bcmvn <- find.pK(sweep.stats)

# 准备数据并选择最佳 pK（合并，避免重复）
bcmvn$pK_num <- as.numeric(as.character(bcmvn$pK))
best.pK <- bcmvn$pK_num[which.max(bcmvn$BCmetric)]

cat(paste0("Optimal pK: ", best.pK, "\n"))

# 创建 ggplot
p1 <- ggplot(bcmvn, aes(x = pK_num, y = BCmetric)) +
  geom_line(color = "steelblue", linewidth = 1) +
  geom_point(aes(color = pK_num == best.pK), size = 3) +
  scale_color_manual(values = c("steelblue", "red"), guide = "none") +
  geom_vline(xintercept = best.pK, linetype = "dashed", color = "red") +
  labs(
    title = "DoubletFinder Optimal pK Selection",
    subtitle = paste("Optimal pK =", best.pK, 
                     "| Max BCmetric =", round(max(bcmvn$BCmetric), 4)),
    x = "pK",
    y = "BCmetric"
  ) +
  theme_minimal(base_size = 14) +
  theme(plot.title = element_text(face = "bold", hjust = 0.5),
        plot.subtitle = element_text(color = "red", hjust = 0.5))

ggsave("DoubletFinder_pK_selection.pdf", p1, width = 8, height = 6, device = "pdf")

# 继续 DoubletFinder 分析
nExp <- round(doublet.rate * ncol(scRNA))

scRNA <- doubletFinder_v3(
  scRNA,
  PCs = 1:dim.usage,
  pN = 0.25,
  pK = best.pK,
  nExp = nExp,
  reuse.pANN = FALSE,
  sct = FALSE
)

df.col <- grep("DF.classifications", colnames(scRNA@meta.data), value = TRUE)
scRNA$Doublet <- scRNA@meta.data[[df.col]]

cat("Doublet summary:\n")
print(table(scRNA$Doublet))

# ============================================
# 7.1. 在删除RNA里双胞，展示双胞状态
# ============================================
df.col <- grep("DF.classifications", colnames(scRNA@meta.data), value = TRUE)
scRNA$Doublet <- scRNA@meta.data[[df.col]]

cat("双胞过滤前的分类统计:\n")
print(table(scRNA$Doublet))


# ============================================
# 7.2. 过滤掉双胞（只保留Singlet）
# ============================================
cat("\n正在过滤双胞...\n")
scRNA_singlet <- subset(scRNA, subset = Doublet == "Singlet")

# 查看过滤后的细胞数
cat("\n过滤后细胞数:", ncol(scRNA_singlet), "\n")
cat("移除的双胞数:", ncol(scRNA) - ncol(scRNA_singlet), "\n")
cat("保留比例:", round(ncol(scRNA_singlet)/ncol(scRNA)*100, 2), "%\n")

# ============================================
# 7.2.1 重新降维和聚类（因为subset会丢失图结构）
# ============================================
cat("\n重新进行降维和聚类...\n")
scRNA_singlet <- NormalizeData(scRNA_singlet)
scRNA_singlet <- FindVariableFeatures(scRNA_singlet, nfeatures = 2000)
scRNA_singlet <- ScaleData(scRNA_singlet)
scRNA_singlet <- RunPCA(scRNA_singlet, npcs = max(dim.usage, 30))
scRNA_singlet <- FindNeighbors(scRNA_singlet, dims = 1:dim.usage)
scRNA_singlet <- FindClusters(scRNA_singlet, resolution = res.usage)
scRNA_singlet <- RunUMAP(scRNA_singlet, dims = 1:dim.usage)
cat("✓ 降维和聚类完成\n")

# ============================================
# 7.3. 画VlnPlot（使用过滤后的数据）
# ============================================
pdf(paste0("QC_VlnPlot_afterQC_", prefix, ".pdf"), width = 18, height = 5)

VlnPlot(
  scRNA_singlet,
  features = c("nCount_RNA", "nFeature_RNA", "percent.mt", "percent.C"),
  group.by = "Stage",
  pt.size = 0,
  ncol = 4
)

dev.off()
cat(paste0("\n✓ 图片已保存: QC_VlnPlot_afterQC_", prefix, ".pdf\n"))

## =========================
## 8. 去双胞及过滤后 QC 图（使用过滤后的数据）
## =========================
pdf(paste0("UMAP_by_cluster_", prefix, ".pdf"), width = 6, height = 5)
DimPlot(scRNA_singlet, reduction = "umap", label = TRUE)  # ✓ 使用过滤后的数据
dev.off()

pdf(paste0("UMAP_by_Batch_", prefix, ".pdf"), width = 6, height = 5)
DimPlot(scRNA_singlet, reduction = "umap", group.by = "Batch")  # ✓ 使用过滤后的数据
dev.off()
 
pdf(paste0("UMAP_by_Batch_split_", prefix, ".pdf"), width = 18, height = 5)
DimPlot(scRNA_singlet, reduction = "umap", group.by = "Batch", split.by = "Batch", ncol = 3)  # ✓ 使用过滤后的数据
dev.off()

# ============================================
# 9. 保存过滤双胞后的RDS（使用 prefix 变量）
# ============================================
save_path <- paste0("/data/work/RNA_AraSeed/rds/", prefix, "_after_QC_doublet_filtered.rds")
saveRDS(scRNA_singlet, save_path)
cat("✓ RDS已保存:", save_path, "\n")
```
