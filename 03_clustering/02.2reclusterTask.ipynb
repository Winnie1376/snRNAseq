```r
library(Seurat)
library(ggplot2)
library(cowplot)
library(Matrix)
library(dplyr)
library(patchwork)
library(clustree, lib.loc = "/data/work/renv/library/R-4.2/x86_64-conda-linux-gnu")
library(data.table, lib.loc = "/opt/conda/lib/R/library")

# RDS 文件和 SampleID
rds_file <- "/data/work/RNA_AraSeed/rds/cot/cot_after_QC_doublet_filtered.rds"  #🥰
SampleID <- "cot_recluster" #🥰

# 输出路径
output_dir <- "/data/work/RNA_AraSeed/rds/cot" #🥰

# 参数设置
dim.usage <- as.numeric(20)
res.usage <- as.numeric(0.9)

# 读取数据
scRNA <- readRDS(rds_file)

## =========================
## 标准降维与聚类
## =========================

scRNA <- NormalizeData(scRNA)
scRNA <- FindVariableFeatures(scRNA, nfeatures = 2000)
scRNA <- ScaleData(scRNA)
scRNA <- RunPCA(scRNA, npcs = max(dim.usage, 30))

scRNA <- FindNeighbors(scRNA, dims = 1:dim.usage)
scRNA <- FindClusters(scRNA, resolution = res.usage)
scRNA <- RunUMAP(scRNA, dims = 1:dim.usage)

# 保存UMAP图
pdf(file.path(output_dir, paste0("UMAP_by_cluster_", SampleID, ".pdf")), width = 6, height = 5)
DimPlot(scRNA, reduction = "umap", label = TRUE)
dev.off()

pdf(file.path(output_dir, paste0("UMAP_by_Batch_", SampleID, ".pdf")), width = 6, height = 5)
DimPlot(scRNA, reduction = "umap", group.by = "Batch")
dev.off()

pdf(file.path(output_dir, paste0("UMAP_by_Batch_split_", SampleID, ".pdf")), width = 18, height = 5)
DimPlot(scRNA, reduction = "umap", group.by = "seurat_clusters", split.by = "Batch", ncol = 3)
dev.off()

# 保存重新聚类后的RDS
saveRDS(scRNA, file.path(output_dir, paste0(SampleID, ".rds")))
cat("✓ 重新聚类完成，RDS已保存\n")
```
